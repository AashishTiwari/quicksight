## Sentiment Analysis

For sentiment analysis our key resources are SEC filings and Quarterly Earnings Transcripts, these are mined for information and theme analysis as well.
We use the CxO level commentary and answers, this includes CEOs, CFOs, CDOs, President, Investor Relations etc.

For the purpose of Sentiment output, we will consider analysis for two companies, 
- Vertex INC.  
- Bed Bath & Beyond Inc.

From the sentiment graphs, we can observe that Vertex had some problems in the last 3 quarters before 2021Q2, in the recent quarter things have stabilized and positive acceleration can be seen.
This is evident and proved by Analysts questions in 2021Q2 call, one of the analyst can be heard saying :

<blockquote> "Great, thank you. And congratulations on the reacceleration in the business. Hey, David, can I just go really big picture to start us out, which is you had five quarters where the business is decelerating. And now you have this really nice reacceleration. And during the period of the acceleration, one of the things that you were pointing to was, hey, if people aren't implementing ERP, that is a headwind for us. How has that environment changed? " </blockquote> 

For Bed Bath & Beyond (BBBY), the news and other sources clearly indicate that the company is struggling to perform well and may file for bankruptcy / insolvency if the business continues to decline.
This is further evident from the content analysis, where we can see other sentiments dominating the positive sentiment.

While choosing the companies to approach, we can take a careful view and decide to de-prioritize some struggle companies since they may not be in position for a tech investment.
